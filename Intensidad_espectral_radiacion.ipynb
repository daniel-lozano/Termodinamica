```python
h=6.626*10**-34
c=3*10**8
Kb=1.38*10**-23
print h,c,Kb
```

```python
def B(T,L):
    B=2*(c**2)*h/((L**5)*(e**(h*c/(Kb*T*L))-1))
    return B
```

```python
lam=linspace(0,1,10**6)
#temperaturas ambiente
T1=273
Bl1=zeros(len(lam),dtype=float)

T2=300
Bl2=zeros(len(lam),dtype=float)

T3=400
Bl3=zeros(len(lam),dtype=float)

#temperaturas bajas

T7=1.2
Bl7=zeros(len(lam),dtype=float)

T8=20
Bl8=zeros(len(lam),dtype=float)

T9=100
Bl9=zeros(len(lam),dtype=float)


print lam
```

```python
for i in range(1,len(lam)/100):
    Bl1[i]=B(T1,lam[i])
    Bl2[i]=B(T2,lam[i])
    Bl3[i]=B(T3,lam[i])
    Bl7[i]=B(T7,lam[i])
    Bl8[i]=B(T8,lam[i])
    Bl9[i]=B(T9,lam[i])
```

# Temperaturas ambientes

```python
plot(lam,Bl1,label="T="+str(T1))
plot(lam,Bl2,label="T="+str(T2))
plot(lam,Bl3,label="T="+str(T3))
xlim(min(lam),10**-4)
title("Intensidad espectral de radiacion",size=15)
xlabel("longitud de onda",size=10)
ylabel("Bl",size=15,rotation="horizontal")
legend()
savefig("templado.png")
```
